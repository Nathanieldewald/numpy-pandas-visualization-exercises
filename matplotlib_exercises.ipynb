```python
import numpy as np
import matplotlib.pyplot as plt

# First plot (Square root)
x1 = range(1, 100)
y1 = np.sqrt(x1)

# Second plot (Cubed)
x2 = range(-100, 100)
y2 = [n ** 3 for n in x2]

# Third plot (Exponential)
x3 = range(-100, 100)
y3 = [2 ** n for n in x3]

# Fourth plot (Rational function)
x4 = np.linspace(-10, 10, 100)
y4 = 1 / (x4 + 1)

# Create subplots
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# Plot the data on each subplot
ax[0, 0].plot(x1, y1)
ax[0, 0].set_title("Square Root")

ax[0, 1].plot(x2, y2)
ax[0, 1].set_title("Cubed")

ax[1, 0].plot(x3, y3)
ax[1, 0].set_title("Exponential")

ax[1, 1].plot(x4, y4)
ax[1, 1].set_title("Rational Function")

# Show the combined figure
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# First plot (Square root)
x1 = range(1, 5)
y1 = np.sqrt(x1)

# Second plot (Cubed)
x2 = range(-10, 10)
y2 = [n ** 3 for n in x2]

# Third plot (Exponential)
x3 = range(-10, 10)
y3 = [2 ** n for n in x3]

# Fourth plot (Rational function)
x4 = np.linspace(-1, 1, 10)
y4 = 1 / (x4 + 1)

# Create a single figure
fig, ax = plt.subplots(figsize=(12, 8))

# Plot the data with different colors
ax.plot(x1, y1, label="Square Root", color="blue")
ax.plot(x2, y2, label="Cubed", color="green")
ax.plot(x3, y3, label="Exponential", color="red")
ax.plot(x4, y4, label="Rational Function", color="purple")

# Set title and legend
ax.set_title("Combined Plot of Four Functions")
ax.legend()

# Show the combined figure
plt.show()

```
