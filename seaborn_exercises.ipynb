Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
import env
def get_db_url(database):
    database = database
    url = f'mysql+pymysql://{env.username}:{env.password}@{env.host}/{database}'
    return url

df = pd.read_sql('SELECT * FROM orders', get_db_url('chipotle'))
df['item_price'] = df.item_price.str.replace('$', '').astype(float)
df['revenue'] = df.item_price
df = df.groupby('item_name').quantity.sum().sort_values(ascending=False).head(4)
df
sns.barplot(x=df.index, y=df.values)
#sns.barplot(x='item_name', y='revenue', data=df)
#df.groupby('item_name').revenue.sum().sort_values(ascending=False).head(4).plot.bar(title='Most Popular Items', ylabel='Revenue', xlabel='Item Name')
```


Load the sleepstudy dataset from pydataset and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
from matplotlib.pyplot import subplot

df = data('sleepstudy')
df
sns.lineplot(x='Days', y='Reaction',hue='Subject', data=df, palette='tab10')
average = df.groupby('Days').Reaction.mean()
sns.lineplot(x=average.index, y=average.values, data=average)
```

```python

from matplotlib import pyplot as plt

df = data('sleepstudy')
fig, ax = plt.subplots()

sns.lineplot(x='Days', y='Reaction', hue='Subject', data=df, palette='tab10', ax=ax)
average = df.groupby('Days').Reaction.mean()
sns.lineplot(x=average.index, y=average.values, data=average, color="blue", linewidth=5.5,    ax=ax)

plt.show()
```
